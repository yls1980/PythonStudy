```python
import pandas as pd
data = pd.read_csv('/home/zorro/VM/BigData/mongo/train.csv')
del data['PassengerId']
del data['Name']
```

```python
data_test = pd.read_csv('/home/zorro/VM/BigData/mongo/test.csv')
del data_test['PassengerId']
del data_test['Name']
```

```python
for index, row in data.iterrows():
    train_list.append({ 
            'survival':row['Survived'],
            'pclass':row['Pclass'],
            'sex':row['Sex'],
            'Age':row['Age'],
            'sibsp':row['SibSp'],
            'parch':row['Parch'],
            'ticket':row['Ticket'],
            'fare':row['Fare'],
            'cabin':row['Cabin'],
            'embarked2':row['Embarked'],
      })    
```

```python
for index, row in data.iterrows():
    test_list.append({ 
            'survival':row['Survived'],
            'pclass':row['Pclass'],
            'sex':row['Sex'],
            'Age':row['Age'],
            'sibsp':row['SibSp'],
            'parch':row['Parch'],
            'ticket':row['Ticket'],
            'fare':row['Fare'],
            'cabin':row['Cabin'],
            'embarked2':row['Embarked'],
      })    
```

```python
insert_test_col = test_col.insert_many(test_list)
insert_train_col = train_col.insert_many(train_list)
```
